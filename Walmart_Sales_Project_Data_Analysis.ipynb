# Walmart Sales Data Analysis Project

This notebook analyzes Walmart sales data to identify trends, patterns, and customer behavior using Python libraries like pandas, matplotlib, and seaborn.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Load the Dataset

````python
df = pd.read_csv("WalmartSalesData.csv")
````

## Parse Dates and Extract Time Features
We convert the 'Date' and 'Time' columns into datetime formats so we can extract useful time-based features for analysis.
```python

````python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day_name()
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
````

## Data Overview

# Basic information

````python
df.info()
print(df.describe())
print(df.isnull().sum())
````

## 📈 Total Sales Over Time
This line chart shows how total sales vary each day.
```python

````python
daily_sales = df.groupby('Date')['Total'].sum()
plt.figure(figsize=(12, 6))
daily_sales.plot()
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
````

## ⏰ Sales by Hour of the Day
This chart reveals which times of day generate the most revenue.


````python
plt.figure(figsize=(10, 5))
df.groupby('Hour')['Total'].sum().plot(kind='line', marker='o')
plt.title("Total Sales by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
````

## 🧾 Revenue by Product Line
This bar chart highlights which product lines bring in the most money.

````python
plt.figure(figsize=(10, 5))
df.groupby('Product line')['Total'].sum().sort_values().plot(kind='bar')
plt.title("Revenue by Product Line")
plt.ylabel("Total Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

## 🌟 Average Rating by Product Line
This chart displays how customers rated each product line on average.

````python
plt.figure(figsize=(10, 5))
df.groupby('Product line')['Rating'].mean().sort_values().plot(kind='bar')
plt.title("Average Rating by Product Line")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

##  Gender Distribution
Pie chart showing the gender split of customers.

````python
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Customer Gender Distribution")
plt.ylabel("")
plt.tight_layout()
plt.show()
````

## Payment Method Usage
This chart breaks down the popularity of each payment method.

````python
plt.figure(figsize=(8, 5))
df['Payment'].value_counts().plot(kind='bar')
plt.title("Payment Method Usage")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
````
